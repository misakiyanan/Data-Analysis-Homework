```python
import pandas as pd
import pymysql
```

```python
author_act = pd.read_csv('author_act.csv')
author_act
```

```python
def date_formatting(date_str):
    d, m, y = date_str.split('/')[0],  date_str.split('/')[1],  date_str.split('/')[2]
    if len(d) == 1:
        d = "0" + d
    if len(m) == 1:
        m = "0" + m
    formatted = '20'+y+'-'+m+'-'+d
    return formatted
```

```python
date_formatting('6/7/21')
```

```python
author_act['dates'] = author_act['dates'].apply(date_formatting)
```

```python
records = author_act.to_records(index=False).tolist()
records  # with datetime format revised
```
